**Here is the name of the fils that included in Dataset.**

```python
files = [
    "CAvideos.csv",
    "DEvideos.csv",
    "FRvideos.csv",
    "GBvideos.csv",
    "INvideos.csv",
    "JPvideos.csv",
    "KRvideos.csv",
    "MXvideos.csv",
    "RUvideos.csv",
    "USvideos.csv"
]
```

**1. Create a single dataframe with the concatenation of all input csv files, adding a column called country**

```python
overall = []
```

```python
for file in files:
    
    df_temp = pd.read_csv("../data/" + file, encoding="latin1")
    
    country = file[:2]
    df_temp["country"] = country
    
    overall.append(df_temp)
    
    
df = pd.concat(overall, ignore_index= True)  
    
```
