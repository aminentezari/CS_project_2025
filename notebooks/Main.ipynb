```python
files = [
    "CAvideos.csv",
    "DEvideos.csv",
    "FRvideos.csv",
    "GBvideos.csv",
    "INvideos.csv",
    "JPvideos.csv",
    "KRvideos.csv",
    "MXvideos.csv",
    "RUvideos.csv",
    "USvideos.csv"
]
```

**1. Create a single dataframe with the concatenation of all input csv files, adding a column called country**

```python
overall = []
```

```python
for file in files:
    
    df_temp = pd.read_csv("../data/" + file, encoding="utf-8",encoding_errors="replace")
    
    country = file[:2]
    
    df_temp["country"] = country
    
    overall.append(df_temp)
    
    
df = pd.concat(overall, ignore_index= True)  
    
```

### This is done by string but there is way also for the converting into datetime

```python
df["trending_date"] = pd.to_datetime(df["trending_date"], format="%y.%d.%m")
```

```python
df["year"] = df["trending_date"].dt.year
df["month"] = df["trending_date"].dt.month
df["day"] = df["trending_date"].dt.day
```
